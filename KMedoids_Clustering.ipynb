```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn_extra.cluster import KMedoids
from sklearn.datasets import load_digits
from sklearn.decomposition import PCA
from sklearn.preprocessing import scale
```

```python
data = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/00267/data_banknote_authentication.txt")
data
```

```python
data.columns = ['VWT','SWT','CWT','E','C']
```

```python
df = data.drop(['E'],axis=1) 
df.head()
```

```python
from sklearn import preprocessing

x = df.values #returns a numpy array
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
normalized_df = pd.DataFrame(x_scaled)
normalized_df.columns = ['VWT','SWT','CWT','C']

print(normalized_df)
```

```python
df = normalized_df
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

z_axis = df['SWT']
x_axis = df['VWT']
y_axis = df['CWT']
ax.scatter3D(x_axis, y_axis, z_axis, c=df['C'], cmap='rainbow')
```

```python
sns.set(style="white", palette="muted", color_codes=True)


f, axes = plt.subplots(1, 3, figsize=(18, 7), sharex=True)
sns.despine(left=True)

sns.scatterplot(df['SWT'],df['VWT'], hue=df['C'], cmap='rainbow', ax=axes[0])
sns.scatterplot(df['SWT'],df['CWT'], hue=df['C'], cmap='rainbow', ax=axes[1])
sns.scatterplot(df['VWT'],df['CWT'], hue=df['C'], cmap='rainbow', ax=axes[2])


plt.setp(axes, yticks=[])
plt.tight_layout()
```

```python
kmedoids = KMedoids(metric="manhattan", n_clusters=5, init="heuristic", max_iter=2).fit(df)
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

z_axis = df['VWT']
x_axis = df['SWT']
y_axis = df['CWT']
ax.scatter3D(x_axis, y_axis, z_axis, c=kmedoids.labels_, cmap='rainbow' )
```

```python
sns.set(style="white", palette="muted", color_codes=True)


f, axes = plt.subplots(1, 3, figsize=(19, 7), sharex=True)
sns.despine(left=True)

sns.scatterplot(df['SWT'],df['VWT'], hue=kmedoids.labels_, cmap='rainbow', style=df['C'], markers = {0: "X", 1: "s"}, ax=axes[0])
sns.scatterplot(df['SWT'],df['CWT'], hue=kmedoids.labels_, cmap='rainbow', style=df['C'], markers = {0: "X", 1: "s"}, ax=axes[1])
sns.scatterplot(df['VWT'],df['CWT'], hue=kmedoids.labels_, cmap='rainbow', style=df['C'], markers = {0: "X", 1: "s"}, ax=axes[2])

plt.setp(axes, yticks=[])
plt.tight_layout()
```

```python
models = [
     (
         KMedoids(metric="manhattan", n_clusters=4, 
         init="heuristic"),"Manhattan metric",
     ),
     (
         KMedoids(metric="euclidean", n_clusters=4,  
         init="heuristic"),"Euclidean metric",
     ),
     (
         KMedoids(metric="cosine", n_clusters=4,
         init="heuristic"), "Cosine metric", 
     ),
 ] 
```

```python
plt.clf()
plt.figure(figsize=(15,10))

for i, (model, description) in enumerate(models):

      kmedoid = model.fit(df)
      fig = plt.figure()
      ax = plt.axes(projection='3d')

      z_axis = df['VWT']
      x_axis = df['SWT']
      y_axis = df['CWT']
      ax.scatter3D(x_axis, y_axis, z_axis, c=kmedoids.labels_, cmap='rainbow' )

      plt.title(description)

      sns.set(style="white", palette="muted", color_codes=True)


      f, axes = plt.subplots(1, 3, figsize=(19, 7), sharex=True)
      sns.despine(left=True)

      sns.scatterplot(df['SWT'],df['VWT'], hue=kmedoids.labels_, cmap='rainbow', style=df['C'], markers = {0: "X", 1: "s"}, ax=axes[0])
      sns.scatterplot(df['SWT'],df['CWT'], hue=kmedoids.labels_, cmap='rainbow', style=df['C'], markers = {0: "X", 1: "s"}, ax=axes[1]).set(title=description)
      sns.scatterplot(df['VWT'],df['CWT'], hue=kmedoids.labels_, cmap='rainbow', style=df['C'], markers = {0: "X", 1: "s"}, ax=axes[2])
      
      plt.setp(axes, yticks=[])
      
      plt.tight_layout()    

```
