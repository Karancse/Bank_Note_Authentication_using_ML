```python
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
import numpy as np
import scipy.cluster.hierarchy as shc
from sklearn.cluster import AgglomerativeClustering
import seaborn as sns
sns.set()
```

```python
data = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/00267/data_banknote_authentication.txt")
data
```

```python
data.columns = ['VWT','SWT','CWT','E','C']
df = data.drop(['E'],axis=1) 
df.head()
```

```python
from sklearn import preprocessing

x = df.values #returns a numpy array
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
normalized_df = pd.DataFrame(x_scaled)
normalized_df.columns = ['VWT','SWT','CWT','C']

print(normalized_df)
```

```python
df = normalized_df
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

z_axis = df['SWT']
x_axis = df['VWT']
y_axis = df['CWT']
ax.scatter3D(x_axis, y_axis, z_axis, c=df['C'], cmap='rainbow')
```

```python
plt.figure(figsize=(10, 7))
plt.title("Customer Dendograms")
dend = shc.dendrogram(shc.linkage(df, method='ward'))
```

```python
cluster = AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='ward')
cluster.fit_predict(df)
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
z_axis = df['SWT']
x_axis = df['VWT']
y_axis = df['CWT']
ax.scatter3D(x_axis, y_axis, z_axis, c=cluster.labels_)
```

```python
sns.set(style="white", palette="muted", color_codes=True)


f, axes = plt.subplots(1, 3, figsize=(19, 7), sharex=True)
sns.despine(left=True)

sns.scatterplot(df['SWT'],df['VWT'], hue=cluster.labels_, cmap='rainbow', style=df['C'], markers = {0: "X", 1: "s"}, ax=axes[0])
sns.scatterplot(df['SWT'],df['CWT'], hue=cluster.labels_, cmap='rainbow', style=df['C'], markers = {0: "X", 1: "s"}, ax=axes[1])
sns.scatterplot(df['VWT'],df['CWT'], hue=cluster.labels_, cmap='rainbow', style=df['C'], markers = {0: "X", 1: "s"}, ax=axes[2])

plt.setp(axes, yticks=[])
plt.tight_layout()

"""

f, axes = plt.subplots(1, 3, figsize=(18, 7), sharex=True)
sns.despine(left=True)

sns.scatterplot(df['SWT'],df['VWT'], hue=cluster.labels_, style=df['C'], size=1 , cmap='rainbow', markers = {0: "s", 1: "X"} , palette="deep", ax=axes[0])
sns.scatterplot(df['SWT'],df['CWT'], hue=cluster.labels_, style=df['C'], cmap='rainbow', markers = {0: "o", 1: "X"} , palette="deep", ax=axes[1])
sns.scatterplot(df['VWT'],df['CWT'], hue=cluster.labels_, style=df['C'], cmap='rainbow', ax=axes[2])

plt.setp(axes, yticks=[])
plt.tight_layout()

"""
```
